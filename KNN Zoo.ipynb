```python
##Creating dummy variables for the animalname variable since it is a categorical data 
dummy = pd.get_dummies(zoo["animalname"],drop_first =True)
dummy
```

```python
## concatinate back dummy animal into zoo data set. Now the data can be used since it is now binary.

zoo = pd.concat([zoo,dummy],axis=1)
zoo = zoo.drop(["animalname"],axis=1)
zoo
```

```python
## factorizing "legs" to make it as a categorical.ie 2-legged, 4-legged, 8-legged
zoo["legs"].value_counts()
zoo["legs"],_=pd.factorize(zoo["legs"])
```

```python
zoo.legs
```

```python
zoo.describe()
```

```python
features = zoo.drop(["type"],axis=1)
features
```

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
# Normalized data frame (considering the numerical part of data) normalizing all rows : 0-9 columns
features_n = norm_func(features)
features_n
```

```python
X = np.array(features_n) # Predictors. Taken from normalized dataset
Y = np.array(zoo['type']) # Target. Taken from original dataset
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)    #30% spilt test

from sklearn.neighbors import KNeighborsClassifier
```

```python
## fitting the data then predicting the data. n start with odd numbers
model1 = KNeighborsClassifier(n_neighbors=5)
model1.fit(X_train, Y_train)
```

```python
##predicting model1 (train)
pred1 = model1.predict(X_train)
pred1
```

```python
##Finding the accuracy of the model1 on training data
train_accuracy = np.mean(model1.predict(X_train)==Y_train)
train_accuracy
```

Train accuracy for model1 is 94.28%

```python
##predicting model1 (test)
pred1 = model1.predict(X_test)
pred1
```

```python
##Accuracy on model1 test data
test_accuracy = np.mean(model1.predict(X_test)==Y_test)
test_accuracy
```

Test accuracy for model1 is 87.1%

```python
##Changing the K value
model2 = KNeighborsClassifier(n_neighbors=9)
model2.fit(X_train, Y_train)
```

```python
##predicting model2 (train)
pred2 = model2.predict(X_train)
pred2
```

```python
##Finding the accuracy of the model2 on training data
train_accuracy = np.mean(model2.predict(X_train)==Y_train)
train_accuracy
```

Train accuracy for model2 is 92.86%

```python
##predicting model2 (test)
pred2 = model2.predict(X_test)
pred2
```

```python
##Accuracy on model2 test data
test_accuracy = np.mean(model2.predict(X_test)==Y_test)
test_accuracy
```

Test accuracy for model2 is 87.1%

```python
# creating empty list variable to store accuracy values
acc = []
```

```python
# running KNN algorithm for 3 to 50 nearest neighbours(odd numbers) and 
# storing the accuracy values 
for i in range (4,30,1):
    model1 = KNeighborsClassifier(n_neighbors = i)
    model1.fit(X_train,Y_train)
    train_acc = np.mean(model1.predict(X_train)==Y_train)
    test_acc = np.mean(model1.predict(X_test)==Y_test)
    acc.append([train_acc, test_acc])
```

```python
acc
```

```python
##Training accuracy plot
plt.plot(np.arange(4,30,1),[i[0] for i in acc],'bo-')
```

```python
##Test accuracy plot

plt.plot(np.arange(4,30,1),[i[1] for i in acc],'ro-')

plt.legend(["train","test"])
```

# Therefore, the best model for zoo animal classification is model1 by using n=5 with 94.28% accuracy in the training data
